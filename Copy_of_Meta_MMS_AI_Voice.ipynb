```python
from ttsmms import TTS
```

```python
tts=TTS("data/mya") #update lang code
```

Short-Text

```python
wav=tts.synthesis("မြန်မာလို ပြောကြည့်ပါ")
```

```python
from IPython.display import Audio
Audio(wav["x"], rate=wav["sampling_rate"])
```

Slightly Long Text

```python
text = """တိမ်ထူသော သို့မဟုတ် မိုးရွာသော ရက်များတွင် ဆိုလာမှ ဓာတ်အား လုံလောက်ထုတ်လုပ်နိုင်ခြင်းမရှိသောကြောင့် ၄င်းရက်အတွင်း လိုအပ်သော လျှပ်စစ်ရရှိနိင်ရန်အတွက် ဘက်ထရီစွမ်းအင် သိုလောင်လိုသည့်ရက်ကို သတ်မှတ်ရန်"""
```

```python
wav_text =tts.synthesis(text)
```

```python
from IPython.display import Audio
Audio(wav_text["x"], rate=wav["sampling_rate"])
```
